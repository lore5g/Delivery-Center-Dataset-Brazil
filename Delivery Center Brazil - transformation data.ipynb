```python
import pandas as pd
import scipy
```

```python
df = pd.read_csv("orders1.csv")
df.head(3)
```

Removing irrelevant Columns and Variables which its meaning is unknown

```python
df = df.drop(["order_moment_delivered", "order_moment_in_expedition", "order_metric_collected_time", "order_metric_paused_time", "order_metric_production_time", "order_metric_walking_time", "order_metric_expediton_speed_time", "order_metric_transit_time", "order_metric_cycle_time"], axis = 1)
df.head(3)
```

Understanding the existing data types

```python
df.info()
```

```python
date_columns = df.iloc[:, 14:21]
for column in date_columns:
    df[column] = pd.to_datetime(df[column])
df["order_moment_finished"] = pd.to_datetime(df["order_moment_finished"])
```

```python
df.info()
```

Removing the Cancelled orders
###### Due to Cancelled orders have multiple missing data and represent only 5% of the dataset, we are removing them

```python
df.order_status.value_counts()
```

```python
df = df.set_index("order_status")
df = df.drop("CANCELED")
df
```

```python
df = df.reset_index()
first_column = df.pop("order_id")
df.insert(0, "order_id", first_column)
df
```

Managing null values

```python
for c in df.iloc[:, 14:22]:
    print(df[c].isnull().value_counts()) #Select all rows with Null values for a specific column
```

```python
df = df.fillna(method='ffill')
```

```python
df.isna().sum()
```
